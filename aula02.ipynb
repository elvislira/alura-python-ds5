# Aula 02 - Utilizando planilhas

```python
import pandas as pd
```

* Lendo arquivo do Excel

```python
url = 'https://github.com/alura-cursos/Pandas/blob/main/emissoes_CO2.xlsx?raw=True'

emissoes_co2 = pd.read_excel(url)
emissoes_co2
```

* Descobrindo o nome das páginas do arquivo do excel

```python
pd.ExcelFile(url).sheet_names
```

* Lendo páginas específicas de uma planilha

```python
percapita = pd.read_excel(url, sheet_name='emissoes_percapita')
percapita.head()
```

```python
fontes = pd.read_excel(url, sheet_name='fontes')
fontes.head()
```

* Lendo colunas específicas

```python
intervalo = pd.read_excel(url, sheet_name='emissoes_C02', usecols='A:D')
intervalo.head()
```

* Restrigindo o número de linhas

```python
intervalo = pd.read_excel(url, sheet_name='emissoes_C02', usecols='A:D', nrows=10)
intervalo
```

* Salvando arquivo no formato excel (xlsx)

```python
percapita.to_excel('dados/CO2_percapta.xlsx', index=False)
```
