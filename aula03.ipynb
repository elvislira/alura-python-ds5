* Lendo dados JOSON hierarquizados

```python
dados_pacientes_2 = pd.read_json('dados_brutos/pacientes_2.json')
dados_pacientes_2
```

* Normalizando os dados hierarquizados

```python
df_normalizado = pd.json_normalize(dados_pacientes_2['Pacientes'])
df_normalizado
```

```python
import json

with open('dados_brutos/pacientes_2.json', 'r') as file:
    dados = json.loads(file.read())

pd.json_normalize(dados, record_path='Pacientes', meta=['Pacientes', 'Ano']).iloc[:,:14]
```

* Obtendo os dados de uma API

```python
import requests
import json

url_api = 'https://fruityvice.com/api/fruit/all'
dados_frutas = requests.get(url_api)
resultado = json.loads(dados_frutas.text)

dados_nutricionais = pd.json_normalize(resultado)
dados_nutricionais
```

```python
url_api_starwars = 'https://swapi.dev/api/'
root = requests.get(url_api_starwars)
resultado = json.loads(root.text)
pd.json_normalize(resultado)
```

```python
url_starships = 'https://swapi.dev/api/starships/'
get_starships = requests.get(url_starships)
resultado_starships = json.loads(get_starships.text)
df = pd.json_normalize(resultado_starships)
pd.json_normalize(df['results'][0])
```
