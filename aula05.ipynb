# Aula 05 - Trabalhando com banco de dados

```python
import sqlalchemy as db
```

* Criando um banco de dados localmente

```python
engine = db.create_engine('sqlite:///bdados/database.sqlite')
```

```python
import pandas as pd
```

```python
dados = pd.read_csv('dados_brutos/clientes_banco.csv')
dados
```

* Inserindo os dados no banco de dados

```python
dados.to_sql('clientes', engine, index=False)
```

* Inspecionando o banco de dados

```python
inspetor = db.inspect(engine)
print(inspetor.get_table_names())
```

* Todos os clientes cuja classificação de renda é Empregado

```python
query = 'SELECT * FROM clientes WHERE Categoria_de_renda="Empregado"'
```

```python
empregados = pd.read_sql(query, engine)
empregados
```

* Salvando a tabela empregados na base de dados

```python
empregados.to_sql('empregados', con=engine, index=False)
```

* Lendo uma tabela inteira

```python
pd.read_sql_table('empregados', con=engine)
```

* Lendo apenas algumas colunas de uma tabela

```python
pd.read_sql_table('empregados', con=engine, columns=['ID_Cliente', 'Grau_escolaridade', 'Rendimento_anual'])
```

* Excluindo um registro

```python
query = 'DELETE FROM clientes WHERE ID_Cliente=5008804'
```

```python
with engine.connect() as connection:
    with connection.begin():
        resultado = connection.execute(db.text(query))
        print(f'Linhas afetadas: {resultado.rowcount}')
```

```python
pd.read_sql_table('clientes', engine)
```

* Atualizando um registro

```python
query = 'UPDATE clientes SET Grau_escolaridade="Ensino superior" WHERE ID_Cliente=5008808'

with engine.connect() as connection:
    with connection.begin():
        resultado = connection.execute(db.text(query))
        print(f'Linhas afetadas: {resultado.rowcount}')
```

```python
pd.read_sql_table('clientes', engine)
```
